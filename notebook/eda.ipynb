# Toxic Comment Classification - EDA Notebook

```python
# 1. Imports and Setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
from wordcloud import WordCloud

%matplotlib inline
sns.set(style="whitegrid")
```

```python
# 2. Load Dataset
df = pd.read_csv('../data/toxic_comments.csv')  # Adjust path as needed
df.head()
```

```python
# 3. Dataset Overview
print(f"Dataset shape: {df.shape}")
df.info()
df.describe()
```

```python
# 4. Check for Nulls and Duplicates
print("Null values per column:\n", df.isnull().sum())
print("\nDuplicate entries:", df.duplicated().sum())
```

```python
# 5. Visualize Class Distribution
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

plt.figure(figsize=(10, 5))
df[label_cols].sum().sort_values(ascending=False).plot(kind="bar", color="salmon")
plt.title("Number of Comments per Class Label")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Check Multi-Label Distribution
df['label_sum'] = df[label_cols].sum(axis=1)

plt.figure(figsize=(8, 4))
sns.countplot(x='label_sum', data=df, palette="muted")
plt.title("Number of Labels per Comment")
plt.xlabel("Number of Labels Assigned")
plt.ylabel("Comment Count")
plt.show()
```

```python
# 7. Comment Length Analysis
df['comment_length'] = df['comment_text'].fillna("").apply(lambda x: len(x.split()))

plt.figure(figsize=(8, 4))
sns.histplot(df['comment_length'], bins=50, kde=True, color="teal")
plt.title("Distribution of Comment Lengths (Words)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()

print("Longest comment length:", df['comment_length'].max())
print("Shortest comment length:", df['comment_length'].min())
```

```python
# 8. Word Cloud for Toxic Comments
def generate_wordcloud(label):
    text = " ".join(df[df[label]==1]['comment_text'].dropna().astype(str))
    wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='Set2').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud for {label} Comments", fontsize=16)
    plt.show()

for label in label_cols:
    generate_wordcloud(label)
```

```python
# 9. Label Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df[label_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Toxic Comment Labels")
plt.show()
```
